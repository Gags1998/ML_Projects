```python
import matplotlib.pyplot as plt
import numpy as np

# Data points from the house price example
x_data = np.array([1, 2, 3])
y_data = np.array([1, 2, 2])

# Regression coefficients calculated: beta0 = 0.67, beta1 = 0.5
beta0 = 0.67 
beta1 = 0.5

# Generate points for the regression line
x_line = np.linspace(0.5, 3.5, 100)
y_line = beta0 + beta1 * x_line

# Plotting the data
plt.scatter(x_data, y_data, color='red', label='Actual Data (House Prices)')
plt.plot(x_line, y_line, color='blue', label=f'OLS Line: $y = {beta0} + {beta1}x$')

# Adding labels and styling
plt.xlabel('Size (sq ft)')
plt.ylabel('Price ($)')
plt.title('Visualizing OLS Matrix Solution')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

```
